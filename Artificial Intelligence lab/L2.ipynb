```python
#list of list
info = [
    ['12','ashik','cse'],
    ['13','abir', 'eee']
]

id = input()
found = False
for i in range (len(info)):
    if info[i][0] == id:
        print(info[i])
        found = True

if not found:
    print('not found')
```

```python
f = open("input.txt")
nums = {}
l = []
for i in f.readlines():
    i = i.strip()
    l.append(i)
for i in l:
    if i in nums:
        nums[i] += 1
    else :
        nums[i] = 1

print(nums)
```

```python
#function

def find(l):
    mx,mn = l[0],l[0]
    for i in l:
        if i > mx:
            mx = i
        if i < mn:
            mn = i
    
    return mx,mn

l = [1,2,3,4,-22,20000]
res = find(l)
print("min : ", res[1], "\nmax : ", res[0])
```

```python
import math
def euclideanDist(l):
    return math.sqrt(((l[0][0]-l[1][0])**2)+((l[0][1]-l[1][1])**2))

list_of_pair = [[1,2],[3,5]]
euclideanDist(list_of_pair)
```
