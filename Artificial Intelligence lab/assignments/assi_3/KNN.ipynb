KNN Classification:

```

from numpy import genfromtxt
import numpy as np
import random
random.seed(0)

data_path = '/content/drive/MyDrive/AI_lab/assignment/assi_3/iris.csv'
my_data = genfromtxt(data_path, delimiter=',') # numpy array n_samples x (feature+1)

data = my_data.tolist() # 2d list

# separate the dataset
Train_set=[]
Val_set=[]
Test_set=[]

for i in range(0,len(data)):
    R = random.random()
    if R>=0 and R<=0.7:
        Train_set.append(data[i])
    elif R>0.7 and R<=0.85:
        Val_set.append(data[i])
    else :
        Test_set.append(data[i])

def euclideanDist(x,y):
    x = x[:len(x)-1]
    y = y[:len(y)-1]
    point1 = np.array(x)
    point2 = np.array(y)
    dist = np.linalg.norm(point1 - point2)
    return dist

def predict_classification(Neighbors):
    Classes = []
    for i in Neighbors:
        Classes.append(i[-1])
    prediction = max(Classes, key= Classes.count)
    return(prediction)
    
#Calculate validation accuracy for K = 1, 3, 5, 10, 15
k = 15
count = 0

for V in Val_set:
    L = []
    for T in Train_set:
        L.append((T,euclideanDist(V,T)))
    L.sort(key = lambda x: x[1])
    L = L[:k] #slice the k nearest Neighbors

    # Using map for 0 index of each tuple of L, at 0 index of each tuple T is exist
    a = map(lambda x: x[0], L)
    Neighbors =  list(a)
    identified_class = predict_classification(Neighbors)

    if identified_class == V[-1] :
        count += 1
    
    # print('validation set ',V)
    # print(L[:k])
    # print('Neighbors ', Neighbors)
    # print(identified_class)
    # print('count ', count)

validation_accuracy = (count / len(Val_set)) * 100

print('validation_accuracy ', validation_accuracy)
```

KNN Regression:


```

from numpy import genfromtxt
import numpy as np
import random
random.seed(0)

data_path = '/content/drive/MyDrive/AI_lab/assignment/assi_3/diabetes.csv'
my_data = genfromtxt(data_path, delimiter=',') # numpy array n_samples x (feature+1)

data = my_data.tolist() # 2d list

# separate the dataset
Train_set=[]
Val_set=[]
Test_set=[]

for i in range(0,len(data)):
    R = random.random()
    if R>=0 and R<=0.7:
        Train_set.append(data[i])
    elif R>0.7 and R<=0.85:
        Val_set.append(data[i])
    else :
        Test_set.append(data[i])

def euclideanDist(x,y):
    x = x[:len(x)-1]
    y = y[:len(y)-1]
    point1 = np.array(x)
    point2 = np.array(y)
    dist = np.linalg.norm(point1 - point2)
    return dist

def average(Neighbors):
    sum = 0
    for i in Neighbors:
        sum += i[-1]
    avg = sum / len(Neighbors)
    return avg
    
#Calculate validation accuracy for K = 1, 3, 5, 10, 15
k = 15
Error = 0

for V in Val_set:
    L = []
    for T in Train_set:
        L.append((T,euclideanDist(V,T)))
    L.sort(key = lambda x: x[1])
    L = L[:k] #slice the k nearest Neighbors

    # Using map for 0 index of each tuple of L, at 0 index of each tuple T is exist
    a = map(lambda x: x[0], L)
    Neighbors =  list(a)
    avg = average(Neighbors)

    Error = Error + ( V[-1] - avg)**2

Mean_Squared_Error = Error / len(Val_set)

print('Mean_Squared_Error ', Mean_Squared_Error)
```
